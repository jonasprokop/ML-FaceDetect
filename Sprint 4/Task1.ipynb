```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = "identity_CelebA.txt" 
df = pd.read_csv(file_path, sep=" ")
df.columns = ['Path', 'Identity']
identity_counts = df.groupby(['Identity'])
df['count'] = df.groupby('Identity').transform('count')
df = df.drop_duplicates(subset = ["Identity"])

```

```python
df.hist(column = "Identity")
```
