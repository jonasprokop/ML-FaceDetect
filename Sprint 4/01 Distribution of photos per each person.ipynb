```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

file_path = "identity_CelebA.txt" 
df = pd.read_csv(file_path, sep=" ", header=None)
df.columns = ['Path', 'Identity']
identity_counts = df.groupby(['Identity'])
df['count'] = df.groupby('Identity').transform('count')
df = df.drop_duplicates(subset = ["Identity"])

```

```python
df.head()
#df.shape
```

```python
histogram_of_count = df.hist(column = "count")
histogram_of_count
```

```python
%%time
x = df["Identity"]
bins = len(df["Identity"])
counts = df["count"]

plt.figure(figsize=(10,5))
histogram_of_identities = plt.hist(bins = bins, x = x, weights = counts)
histogram_of_identities
```
