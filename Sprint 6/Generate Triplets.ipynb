```python
import pandas as pd
import random
import numpy as np
import copy
```

```python
#loading file
file_path = "identity_CelebA.txt" 
df = pd.read_csv(file_path, sep=" ", header=None)
df.columns = ['Path', 'Identity']

#creates a dict of identities with their respective paths
paths_list = {}
for x in range(len(df)):
    id = df.iloc[x, 1]
    if id not in paths_list:
        paths_list[id] = []
    paths_list[id].append(df.iloc[x, 0])
    
#finding identities with <2 paths
identity_counts = df.groupby(['Identity'])
df['count'] = df.groupby('Identity').transform('count')
df = df.drop_duplicates(subset = ["Identity"])
filter = df["count"]<2
df = df.where(filter)
df = df.dropna()
bad_ids = df["Identity"].values.tolist()
bad_ids = [int(v) for v in bad_ids]
```

```python
def create_triplets(data, count):
    triplets = []
    for i in range(count):
        path_list = copy.deepcopy(data)
        ids = list(path_list.keys())
        #removing ids from the bad_ids list
        ids_for_same = [i for i in ids if i not in bad_ids]
        identity = random.choice(ids_for_same)
        reference_path = random.choice(path_list.get(identity))
        #removing reference_path to avoid two same paths
        path_list[identity].remove(reference_path)
        same_path = random.choice(path_list.get(identity))
        #removing identity 
        ids.remove(identity)
        diff_identity = random.choice(ids)
        diff_path = random.choice(path_list.get(diff_identity))
        triplets.append([reference_path, same_path, diff_path])
    return np.array(triplets)
```

```python
%%time
create_triplets(paths_list, 1000)
```
